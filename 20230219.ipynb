### 列出各年度的M2貨幣供給額

#### 方法一

````python
# 方法一:列出各年度的M2貨幣供給額
for i in data:
  print(i["年度"], "==>", i['貨幣供給額期底年增率-M2'])
````

````python
for i in range(len(data)):
  print(i)
````
